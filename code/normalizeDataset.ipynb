```python
import os
import numpy as np
import cv2
```

```python
classes = ['butterfly','cat','chicken','cow','dog','elephant','horse','sheep','spider','squirrel']

#create an already resized and normalized dataset
#https://stackoverflow.com/questions/56569342/detecting-corrupt-images-in-tensorflow
for className in classes:
  os.chdir("c:/Users/thiscomputer/Desktop/animals-10/animals/{}".format(className))
  img_dir = "c:/Users/thiscomputer/Desktop/animals-10/animals/{}".format(className)
  files = os.listdir(img_dir)
  index = -1
  howMany = -1
  for f1 in files:
    index +=1
    if index == howMany: 
      break
    if howMany == -1:
        howMany = 'ALL' 
    print(className,index)  
    img = cv2.imread(f1)
    img = cv2.resize(img,(180,180))
    img = img.astype('float32')
    img /= 255.0
    cv2.imwrite("c:/Users/thiscomputer/Desktop/animals-10/animalsPP{}/{}/{}{}.jpg".format(howMany,className,className,index),img)
```
